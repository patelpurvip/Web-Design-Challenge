# PPatel Homework #11 - Source Weather Data, pandas to_html

```python
import pandas as pd
import numpy as np

raw_weatherdata = "Resources/weatherdata-28.03.20.csv"
weatherdata = pd.read_csv(raw_weatherdata)
```

```python
weatherdata.head()
```

```python
print(weatherdata.to_html())
```

```python
weather_data_html = weatherdata.to_html()

with open("dataframe.weather_data_html", "w", encoding="utf-8") as file:
    file.writelines('<meta charset="UTF-8">\n')
    file.write(weather_data_html)
```
